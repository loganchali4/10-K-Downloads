```
# Import packages
import requests
from bs4 import BeautifulSoup
import os

# Function to donwload 10-K for specified CIK and year
def download_10k(cik, year):
    # IDX URLs
    base_url = "https://www.sec.gov/Archives/edgar/full-index"
    idx_url = f"{base_url}/{year}/QTR"
    
    # Specify headers for connection
    headers = {'User-Agent': 'Logan Chalifour loganchali4@gmail.com'}
    #headers = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}

    # For all four quarters of year
    for quarter in range(1, 5):

        # Create the URL for the IDX file for the given CIK and year
        idx_file_url = f"{idx_url}{quarter}/master.idx"

        # Retrieve the contents of the IDX file and parse it with BeautifulSoup
        idx_response = requests.get(idx_file_url, headers=headers) # Currently blocked from access

        # Write IDC file
        with open('master.idx', 'wb') as f:
            f.write(idx_response.content)

        # Read IDX file
        with open('master.idx', 'r') as f:
            for line in f:
                fields = line.strip('\n').split('|')

                # If the CIK and form type match the criteria for a 10-K filing, download the filing
                if fields[0] == cik and fields[2] == "10-K":
                    # Get the URL of the 10-K filing
                    filing_url = f"https://www.sec.gov/Archives/{fields[4]}"

                    # Send a GET request to the URL and get the response
                    filing_response = requests.get(filing_url, headers=headers)

                    # Get the filename of the 10-K filing
                    filename = fields[1]

                    # Define the path to the local directory where you want to save the file
                    directory = f'{year}'

                    # Create the directory if it doesn't exist
                    if not os.path.exists(directory):
                        os.makedirs(directory)

                    # Save the 10-K filing to the specified directory
                    with open(f'{directory}/{filename}_{year}_10-K.txt', 'wb') as f:
                        # Write the response content to the local file
                        f.write(filing_response.content)

                    # Print a message to indicate that the 10-K has been downloaded
                    print(f"{filename} downloaded to {directory} for CIK {cik} and year {year} QTR{quarter}.")

# Example usage 
download_10k(320193, 2020)
```
